```python
# Importing Required Libraries

import pydriller 
from pydriller import ModificationType
import pandas as pd
import subprocess
from pathlib import Path
```

```python


# Path to the repository
repo_path = Path('./mastodon')

# Check if the repository exists
if repo_path.exists():
    print('Repository exists!')
else:
    # Clone the repository
    subprocess.call(['git', 'clone', 'git://github.com/tootsuite/mastodon.git'])



index_name = "mastodon"
start_tag =  'v3.0.0'
end_tag = 'v4.0.0'
file_type = '.rb'


gitrepo = pydriller.Repository('./mastodon', from_tag=start_tag, to_tag=end_tag)

```

```python

modified_files = []
all_modified_file_names = set()

for commit in gitrepo.traverse_commits():
    modified_files_per_commit = set()
    for file in commit.modified_files:
        if file.filename.endswith(file_type):
            if file.old_path == file.new_path:
                modified_files_per_commit.add(file.filename)
                all_modified_file_names.add(file.filename)
        
    if len(modified_files_per_commit) > 0:
        modified_files.append(modified_files_per_commit)

print(len(modified_files))

```

```python
import numpy as np


all_modified_file_names = list(all_modified_file_names)

filename_indices = []
num_of_changes_per_file = []

for filename in all_modified_file_names:
    indices = []
    num_changes = 0
    
    for index, commit in enumerate(modified_files):
        if filename in commit:
            indices.append(index)
            num_changes += 1
            
    filename_indices.append(indices)
    num_of_changes_per_file.append(num_changes)

# print(all_modified_file_names)
#print(filename_indices)
# print(num_of_changes_per_file)

num_array = np.array(num_of_changes_per_file)
max_indices = num_array.argsort()[-20:][::-1]

most_popular_files = [all_modified_file_names[index] for index in max_indices]
commit_list_per_file = [filename_indices[index] for index in max_indices]

print(most_popular_files)
#print(commit_list_per_file)
```

```python
matrix = dict()
for file in most_popular_files:
  matrix[file] = dict()
  for file_2 in most_popular_files:
    matrix[file][file_2] = 0
```

```python
for commit in modified_files:
  for file in most_popular_files:
    if file in commit: 
      for file_2 in most_popular_files:
        if file_2 in commit:
          matrix[file][file_2] += 1
```
